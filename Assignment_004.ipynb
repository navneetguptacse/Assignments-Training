1. Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** the integers that appeared in **all** three arrays.

    **Example 1:**

    **Input:** arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

    **Output:** [1,5]

    **Explanation:** Only 1 and 5 appeared in the three arrays.

- **C++ Code :**

    ```
    #include <bits/stdc++.h>
    using namespace std;

    vector<int> findCommonElements(vector<int>& arr1, vector<int>& arr2, vector<int>& arr3) {
        vector<int> result;
        int i = 0, j = 0, k = 0;

        while (i < arr1.size() && j < arr2.size() && k < arr3.size()) {
            if (arr1[i] == arr2[j] && arr2[j] == arr3[k]) {
                result.push_back(arr1[i]);
                i++;
                j++;
                k++;
            } else if (arr1[i] < arr2[j]) {
                i++;
            } else if (arr2[j] < arr3[k]) {
                j++;
            } else {
                k++;
            }
        }

        return result;
    }

    int main() {
        int n1, n2, n3;
        cout << "Enter the size of arr1: ";
        cin >> n1;
        cout << "Enter the elements of arr1 (sorted in strictly increasing order): ";
        vector<int> arr1(n1);
        for (int i = 0; i < n1; i++) {
            cin >> arr1[i];
        }

        cout << "Enter the size of arr2: ";
        cin >> n2;
        cout << "Enter the elements of arr2 (sorted in strictly increasing order): ";
        vector<int> arr2(n2);
        for (int i = 0; i < n2; i++) {
            cin >> arr2[i];
        }

        cout << "Enter the size of arr3: ";
        cin >> n3;
        cout << "Enter the elements of arr3 (sorted in strictly increasing order): ";
        vector<int> arr3(n3);
        for (int i = 0; i < n3; i++) {
            cin >> arr3[i];
        }

        vector<int> commonElements = findCommonElements(arr1, arr2, arr3);

        cout << "Common elements in all three arrays: ";
        for (int num : commonElements) {
            cout << num << " ";
        }
        cout << endl;

        return 0;
    }
    ```

2. Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

    - answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2.
    - answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

    **Note:** that the integers in the lists may be returned in **any** order.

    **Example 1:**

    **Input:** nums1 = [1,2,3], nums2 = [2,4,6]

    **Output:** [[1,3],[4,6]]

    **Explanation:**

    For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

    For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

- **C++ Code :**
    ```
    #include <bits/stdc++.h>
    using namespace std;

    vector<vector<int>> findDisjointIntegers(vector<int>& nums1, vector<int>& nums2) {
        unordered_set<int> set1(nums1.begin(), nums1.end());
        unordered_set<int> set2(nums2.begin(), nums2.end());

        vector<int> distinctNums1;
        vector<int> distinctNums2;

        for (int num : nums1) {
            if (set2.find(num) == set2.end()) {
                distinctNums1.push_back(num);
            }
        }

        for (int num : nums2) {
            if (set1.find(num) == set1.end()) {
                distinctNums2.push_back(num);
            }
        }

        return {distinctNums1, distinctNums2};
    }

    int main() {
        int n1, n2;
        cout << "Enter the size of nums1: ";
        cin >> n1;
        cout << "Enter the elements of nums1: ";
        vector<int> nums1(n1);
        for (int i = 0; i < n1; i++) {
            cin >> nums1[i];
        }

        cout << "Enter the size of nums2: ";
        cin >> n2;
        cout << "Enter the elements of nums2: ";
        vector<int> nums2(n2);
        for (int i = 0; i < n2; i++) {
            cin >> nums2[i];
        }

        vector<vector<int>> disjointIntegers = findDisjointIntegers(nums1, nums2);

        cout << "Distinct integers in nums1 not present in nums2: ";
        for (int num : disjointIntegers[0]) {
            cout << num << " ";
        }
        cout << endl;

        cout << "Distinct integers in nums2 not present in nums1: ";
        for (int num : disjointIntegers[1]) {
            cout << num << " ";
        }
        cout << endl;

        return 0;
    }

    ```

3. Given a 2D integer array matrix, return *the **transpose** of* matrix.

    The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

    **Example 1:**

    **Input:** matrix = [[1,2,3],[4,5,6],[7,8,9]]

    **Output:** [[1,4,7],[2,5,8],[3,6,9]]
    
    ![drawing](./Image/iamge_v3.jpg)

- **C++ Code :**
    ```
    #include <bits/stdc++.h>

    using namespace std;

    vector<vector<int>> transposeMatrix(vector<vector<int>>& matrix) {
        int rows = matrix.size();
        int cols = matrix[0].size();

        vector<vector<int>> transpose(cols, vector<int>(rows, 0));

        for (int i = 0; i < rows; i++) {
            for (int j = 0; j < cols; j++) {
                transpose[j][i] = matrix[i][j];
            }
        }

        return transpose;
    }

    int main() {
        int n, m;
        cout << "Enter the number of rows: ";
        cin >> n;
        cout << "Enter the number of columns: ";
        cin >> m;

        cout << "Enter the elements of the matrix:" << endl;
        vector<vector<int>> matrix(n, vector<int>(m));
        for (int i = 0; i < n; i++) {
            for (int j = 0; j < m; j++) {
                cin >> matrix[i][j];
            }
        }

        vector<vector<int>> transpose = transposeMatrix(matrix);

        cout << "Transpose of the matrix:" << endl;
        for (int i = 0; i < transpose.size(); i++) {
            for (int j = 0; j < transpose[0].size(); j++) {
                cout << transpose[i][j] << " ";
            }
            cout << endl;
        }

        return 0;
    }

    ```

4. Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is **maximized**. Return *the maximized sum*.

    **Example 1:**

    **Input:** nums = [1,4,3,2]

    **Output:** 4

    **Explanation:** 

    All possible pairings (ignoring the ordering of elements) are:
    - (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3
    - (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3
    - (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

    So the maximum possible sum is 4.


- **C++ Code :**

    ```
    #include <bits/stdc++.h>
    using namespace std;

    int arrayPairSum(vector<int>& nums) {
        sort(nums.begin(), nums.end());
        int sum = 0;
        for (int i = 0; i < nums.size(); i += 2) {
            sum += nums[i];
        }

        return sum;
    }

    int main() {
        int n;
        cout << "Enter the value of n: ";
        cin >> n;

        vector<int> nums(2 * n);
        cout << "Enter the array elements: ";
        for (int i = 0; i < 2 * n; i++) {
            cin >> nums[i];
        }

        int maxSum = arrayPairSum(nums);

        cout << "Maximum possible sum: " << maxSum << endl;

        return 0;
    }

    ```

    **Example 1:**

    ![drawing](./Image/v2.jpg)

    **Input:** n = 5

    **Output:** 2

- **C++ Code :**
    ```
    #include <iostream>
    #include <cmath>

    using namespace std;

    int arrangeCoins(int n) {

6. Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

    **Example 1:**

    **Input :** nums = [-4,-1,0,3,10]

    **Output :** [0,1,9,16,100]

    **Explanation:** After squaring, the array becomes [16,1,0,9,100].
    After sorting, it becomes [0,1,9,16,100]

- **C++ Code :**
    
    ```
    #include <iostream>
    #include <vector>
    #include <algorithm>

    using namespace std;

    vector<int> sortedSquares(vector<int>& nums) {
        vector<int> squares(nums.size());

        int left = 0;
        int right = nums.size() - 1;
        int index = nums.size() - 1;

        while (left <= right) {
            int squareLeft = nums[left] * nums[left];
            int squareRight = nums[right] * nums[right];

            if (squareLeft > squareRight) {
                squares[index] = squareLeft;
                left++;
            } else {
                squares[index] = squareRight;
                right--;
            }

            index--;
        }

        return squares;
    }

    int main() {
        int n;
        cout << "Enter the size of the array: ";
        cin >> n;

        vector<int> nums(n);
        cout << "Enter the array elements in non-decreasing order: ";
        for (int i = 0; i < n; i++) {
            cin >> nums[i];
        }

        vector<int> squaredArray = sortedSquares(nums);

        cout << "The array of squared numbers in non-decreasing order: ";
        for (int i = 0; i < squaredArray.size(); i++) {
            cout << squaredArray[i] << " ";
        }
        cout << endl;

        return 0;
    }
    ```

7. You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

    Count and return *the number of maximum integers in the matrix after performing all the operations*

    **Example 1:**

    ![drawing](./Image/q4.jpg)

    **Input:** m = 3, n = 3, ops = [[2,2],[3,3]]

    **Output:** 4

    **Explanation:** The maximum integer in M is 2, and there are four of it in M. So return 4.

- **C++ Code :**

    ```
    #include <bits/stdc++.h>
    using namespace std;

    int maxCount(int m, int n, vector<vector<int>>& ops) {
        int minRow = m;
        int minCol = n;

        for (const auto& op : ops) {
            minRow = min(minRow, op[0]);
            minCol = min(minCol, op[1]);
        }

        return minRow * minCol;
    }

    int main() {
        int m, n;
        cout << "Enter the number of rows (m): ";
        cin >> m;
        cout << "Enter the number of columns (n): ";
        cin >> n;

        int numOps;
        cout << "Enter the number of operations: ";
        cin >> numOps;

        vector<vector<int>> ops(numOps, vector<int>(2));
        cout << "Enter the operations (ai, bi): " << endl;
        for (int i = 0; i < numOps; i++) {
            cin >> ops[i][0] >> ops[i][1];
        }

        int maxIntegers = maxCount(m, n, ops);

        cout << "The number of maximum integers in the matrix: " << maxIntegers << endl;

        return 0;
    }

    ```

8. Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

    *Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

    **Example 1:**

    **Input:** nums = [2,5,1,3,4,7], n = 3

    **Output:** [2,3,5,4,1,7]

    **Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

- **C++ Code :**

    ```
    #include <bits/stdc++.h>
    using namespace std;

    vector<int> shuffleArray(const vector<int>& nums, int n) {
        vector<int> result;
        for (int i = 0; i < n; i++) {
            result.push_back(nums[i]);
            result.push_back(nums[i + n]);
        }
        return result;
    }

    int main() {
        int size;
        cout << "Enter the size of the array: ";
        cin >> size;

        vector<int> nums;
        cout << "Enter the elements of the array: ";
        for (int i = 0; i < 2 * size; i++) {
            int num;
            cin >> num;
            nums.push_back(num);
        }

        vector<int> shuffled = shuffleArray(nums, size);

        cout << "Output: ";
        for (int num : shuffled) {
            cout << num << " ";
        }

        return 0;
    }
