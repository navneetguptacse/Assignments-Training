2. Explain Semantic tags in html? And why do we need it?

- Semantic tags in HTML are specific elements that carry meaning and describe the structure of the content within a web page. 
- These tags provide a way to semantically mark up different sections and elements, indicating their purpose and role in the document. 
- Instead of using generic `<div>` or `<span>` elements, semantic tags offer more descriptive and meaningful alternatives.

    There are some commonly used sementic tags-

    - `<header>`: Represents the introductory section of a page or a section within the document.
    - `<nav>`: Defines a section containing navigation links.
    - `<main>`: Represents the primary content of the document.
    - `<article>`: Represents a self-contained composition, such as a blog post or a news article.
    - `<section>`: Defines a thematic grouping of content.
    - `<aside>`: Represents content that is tangentially related to the main content, such as sidebars, pull quotes, or advertisements.
    - `<footer>`: Represents the footer section of a document or a section.

8. What is the difference between `<figure>` tag and `<img>` tag?

The `<figure>` and `<img>` tags are both HTML elements used for displaying images, but they serve different purposes.

- The `<img>` tag is a self-closing tag used specifically for embedding images into an HTML document. It does not require a closing tag and must include a src attribute specifying the path to the image file. It is primarily used to display standalone images within the content of a webpage.
**For Example:**

```
<img src="path_to_image.jpg" alt="Description of the image">
```


- The `<figure>` tag is used to group media content, such as images or videos, along with their associated captions or descriptions. It provides a way to semantically mark up a self-contained piece of content that requires additional context or explanation. The actual image or video is typically nested within the `<figure>` element.

**For Example:**

```
<figure>
  <img src="path_to_image.jpg" alt="Description of the image">
  <figcaption>Caption or description of the image</figcaption>
</figure>

```

9. What’s the difference between html tag and attribute and give example of some global attributes?

| HTML Tag                                      | Attribute                                                    |
|-----------------------------------------------|--------------------------------------------------------------|
| HTML tags define the structure and content     | Attributes provide additional information or properties       |
| Enclosed in angle brackets (< >)               | Added within the opening tag of an HTML element               |
| Consist of an opening tag and a closing tag    | Standalone and do not require a closing tag                   |
| Examples: `<p>`, `<div>`, `<h1>`               | Examples: `class`, `id`, `src`, `href`                        |
| Tags define the type of element                | Attributes specify characteristics or behavior                |

Global attributes are attributes that can be used with most HTML tags.

- `class`: Specifies one or more CSS classes to apply to an element for styling purposes.
- `id`: Specifies a unique identifier for an element.
- `style`: Specifies inline CSS styles to be applied to an element.
- `title`: Provides additional information or a tooltip text for an element.
- `data`-*: Allows custom data attributes to be added to elements to store extra information.
- `aria`-*: Used for accessibility purposes to provide additional information to assistive technologies.
