1. Given two strings s1 and s2, return *the lowest **ASCII** sum of deleted characters to make two strings equal*.

    **Example 1:**

    **Input:** s1 = "sea", s2 = "eat"

    **Output:** 231

    **Explanation:** Deleting "s" from "sea" adds the ASCII value of "s" (115) to the sum.

    Deleting "t" from "eat" adds 116 to the sum.

    At the end, both strings are equal, and 115 + 116 = 231 is the minimum sum possible to achieve this.

- **C++ Code :**
    ```
    #include <iostream>
    #include <vector>
    using namespace std;

    int minimumDeleteSum(string s1, string s2) {
        int m = s1.length();
        int n = s2.length();

        vector<vector<int>> dp(m + 1, vector<int>(n + 1, 0));

        for (int i = 1; i <= m; i++) {
            dp[i][0] = dp[i - 1][0] + int(s1[i - 1]);
        }
        for (int j = 1; j <= n; j++) {
            dp[0][j] = dp[0][j - 1] + int(s2[j - 1]);
        }

        for (int i = 1; i <= m; i++) {
            for (int j = 1; j <= n; j++) {
                if (s1[i - 1] == s2[j - 1]) {
                    dp[i][j] = dp[i - 1][j - 1];
                } else {
                    dp[i][j] = min(dp[i - 1][j] + int(s1[i - 1]), dp[i][j - 1] + int(s2[j - 1]));
                }
            }
        }

        return dp[m][n];
    }

    int main() {
        string s1, s2;

        cout << "Enter the first string: ";
        getline(cin, s1);

        cout << "Enter the second string: ";
        getline(cin, s2);

        int result = minimumDeleteSum(s1, s2);
        cout << result << endl;

        return 0;
    }
    ```


2. Given a string s containing only three types of characters: '(', ')' and '*', return true *if* s *is **valid***.

    The following rules define a **valid** string:

    - Any left parenthesis '(' must have a corresponding right parenthesis ')'.
    - Any right parenthesis ')' must have a corresponding left parenthesis '('.
    - Left parenthesis '(' must go before the corresponding right parenthesis ')'.
    - '*' could be treated as a single right parenthesis ')' or a single left parenthesis '(' or an empty string "".

    **Example 1:**

    **Input:** s = "()"

    **Output:** true

- **C++ Code :**
    ```
    #include <iostream>
    #include <stack>
    using namespace std;

    bool isValid(string s) {
        stack<char> parentheses;

        for (char c : s) {
            if (c == '(' || c == '*') {
                parentheses.push(c);
            } else if (c == ')') {
                if (!parentheses.empty()) {
                    parentheses.pop();
                } else if (!parentheses.empty() && parentheses.top() == '*') {
                    parentheses.pop();
                } else {
                    return false;
                }
            }
        }

        while (!parentheses.empty()) {
            if(parentheses.top() == '(' || parentheses.top() == '*') {
                parentheses.pop();
            } else {
                return false;
            }
        }

        return true;
    }

    int main() {
        string s;

        cout << "Enter the string: ";
        getline(cin, s);

        bool result = isValid(s);
        cout << boolalpha << result << endl;

        return 0;
    }

3. Given two strings word1 and word2, return *the minimum number of **steps** required to make* word1 *and* word2 *the same*.

    In one **step**, you can delete exactly one character in either string.

    **Example 1:**

    **Input:** word1 = "sea", word2 = "eat"

    **Output:** 2

    **Explanation:** You need one step to make "sea" to "ea" and another step to make "eat" to "ea".

- **C++ Code :**
    ```
    #include <iostream>
    #include <vector>
    using namespace std;

    int minDistance(string word1, string word2) {
        int m = word1.length();
        int n = word2.length();

        vector<vector<int>> dp(m + 1, vector<int>(n + 1, 0));

        for (int i = 1; i <= m; i++) {
            dp[i][0] = i;
        }
        for (int j = 1; j <= n; j++) {
            dp[0][j] = j;
        }

        for (int i = 1; i <= m; i++) {
            for (int j = 1; j <= n; j++) {
                if (word1[i - 1] == word2[j - 1]) {
                    dp[i][j] = dp[i - 1][j - 1];
                } else {
                    dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1);
                }
            }
        }

        return dp[m][n];
    }

    int main() {
        string word1, word2;

        cout << "Enter the first word: ";
        getline(cin, word1);

        cout << "Enter the second word: ";
        getline(cin, word2);

        int result = minDistance(word1, word2);
        cout << result << endl;

        return 0;
    }
    ```

4. You need to construct a binary tree from a string consisting of parenthesis and integers.

    The whole input represents a binary tree. It contains an integer followed by zero, one or two pairs of parenthesis. The integer represents the root's value and a pair of parenthesis contains a child binary tree with the same structure.
    You always start to construct the **left** child node of the parent first if it exists.

    **Example 1:**

    <img title="2D Array" alt="" width="300px" src="./84.png">

    **Input:** s = "4(2(3)(1))(6(5))"

    **Output:** [4,2,6,3,1,5]

- **C++ Code :**
    ```
    #include <bits/stdc++.h>
    using namespace std;

    struct TreeNode {
        int val;
        TreeNode* left;
        TreeNode* right;
        TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
    };

    TreeNode* buildTree(string s, int start, int end) {
        if (start > end)
            return nullptr;

        int value = 0;
        int sign = 1;
        int i = start;

        if (s[i] == '-') {
            sign = -1;
            i++;
        }
        while (i <= end && isdigit(s[i])) {
            value = value * 10 + (s[i] - '0');
            i++;
        }
        TreeNode* root = new TreeNode(value * sign);

        if (i <= end && s[i] == '(') {
            int count = 1;
            int leftEnd = i + 1;
            while (leftEnd <= end && count > 0) {
                if (s[leftEnd] == '(')
                    count++;
                else if (s[leftEnd] == ')')
                    count--;
                leftEnd++;
            }
            root->left = buildTree(s, i + 1, leftEnd - 2);

            if (leftEnd <= end && s[leftEnd] == '(') {
                int count = 1;
                int rightStart = leftEnd + 1;
                while (rightStart <= end && count > 0) {
                    if (s[rightStart] == '(')
                        count++;
                    else if (s[rightStart] == ')')
                        count--;
                    rightStart++;
                }
                root->right = buildTree(s, leftEnd + 1, rightStart - 2);
            }
        }

        return root;
    }

    void preorderTraversal(TreeNode* root, vector<int>& result) {
        if (root == nullptr)
            return;
        result.push_back(root->val);
        preorderTraversal(root->left, result);
        preorderTraversal(root->right, result);
    }

    vector<int> constructTreeFromString(string s) {
        int n = s.length();
        TreeNode* root = buildTree(s, 0, n - 1);
        vector<int> result;
        preorderTraversal(root, result);
        return result;
    }

    int main() {
        string s;
        cout << "Enter the string representation of the binary tree: ";
        cin >> s;

        vector<int> result = constructTreeFromString(s);

        cout << "Constructed binary tree: ";
        for (int i = 0; i < result.size(); i++) {
            cout << result[i] << " ";
        }
        cout << endl;

        return 0;
    }
    ```

6. Given two strings s and p, return *an array of all the start indices of* p*'s anagrams in* s. You may return the answer in **any order**.

    An **Anagram** is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

    **Example 1:**

    **Input:** s = "cbaebabacd", p = "abc"

    **Output:** [0,6]

    **Explanation:**

    The substring with start index = 0 is "cba", which is an anagram of "abc".

    The substring with start index = 6 is "bac", which is an anagram of "abc".

- **C++ Code :**
    ```
    #include <bits/stdc++.h>
    using namespace std;

    vector<int> findAnagrams(string s, string p) {
        vector<int> result;
        int n = s.length();
        int m = p.length();
        
        if (n < m)
            return result;
        
        unordered_map<char, int> freqP, freqS;
        
        for (char c : p)
            freqP[c]++;
        
        int left = 0, right = 0;
        
        while (right < n) {
            freqS[s[right]]++;
            
            if (right - left + 1 == m) {
                if (freqP == freqS)
                    result.push_back(left);
                
                freqS[s[left]]--;
                if (freqS[s[left]] == 0)
                    freqS.erase(s[left]);
                
                left++;
            }
            
            right++;
        }
        
        return result;
    }

    int main() {
        string s, p;
        cin >> s >> p;

        vector<int> indices = findAnagrams(s, p);

        }

    }

7. Given an encoded string, return its decoded string.

    The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

    You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

    The test cases are generated so that the length of the output will never exceed 105.

    **Example 1:**

    **Input:** s = "3[a]2[bc]"

    **Output:** "aaabcbc"

- **C++ Code :**
    ```
    #include <bits/stdc++.h>
    using namespace std;

    string decodeString(string s) {
        stack<string> st;
        for (char c : s) {
            if (c == ']') {
                string decoded_str = "";
                while (st.top() != "[") {
                    decoded_str = st.top() + decoded_str;
                    st.pop();
                }
                st.pop();  // Pop the opening bracket
                int count = stoi(st.top());
                st.pop();
                string repeated_str = "";
                while (count > 0) {
                    repeated_str += decoded_str;
                    count--;
                }
                st.push(repeated_str);
            }
            else {
                string str(1, c);
                st.push(str);
            }
        }
        string decoded = "";
        while (!st.empty()) {
            decoded = st.top() + decoded;
            st.pop();
        }
        return decoded;
    }

8. Given two strings s and goal, return true *if you can swap two letters in* s *so the result is equal to *goal*, otherwise, return *false*.*

    Swapping letters is defined as taking two indices i and j (0-indexed) such that i != j and swapping the characters at s[i] and s[j].

    - For example, swapping at indices 0 and 2 in "abcd" results in "cbad".

    **Example 1:**

    **Input:** s = "ab", goal = "ba"

    **Output:** true

    **Explanation:** You can swap s[0] = 'a' and s[1] = 'b' to get "ba", which is equal to goal.

- **C++ Code :**
    ```
    #include <bits/stdc++.h>
    using namespace std;

    bool canSwap(string s, string goal) {
        vector<int> diff_s, diff_goal;

        for (int i = 0; i < s.length(); i++) {
            if (s[i] != goal[i]) {
                diff_s.push_back(i);
                diff_goal.push_back(i);
            }
        }

        if (diff_s.size() != 2 || diff_goal.size() != 2) {
            return false;
        }

        swap(s[diff_s[0]], s[diff_s[1]]);

        return s == goal;
    }

    int main() {
        string s, goal;
        cout << "Enter string s: ";
        cin >> s;
        cout << "Enter string goal: ";
        cin >> goal;

        bool result = canSwap(s, goal);
        cout << (result ? "true" : "false") << endl;

        return 0;
    }
